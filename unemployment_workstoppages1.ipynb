```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('workStoppage.csv')

df.plot(x='Year', y='Value') 
plt.xticks(rotation=45)
plt.title('Workers Affected by Work Stoppages')
plt.xlabel('Year')
plt.ylabel('Workers Affected (in Thousands)')
plt.figure(figsize=(10,6))
plt.show()
                   
df2 = pd.read_csv('unemployment.csv')

df2.plot(x='Label', y='Value') 
plt.xticks(rotation=45)
plt.title('Unemployment Rate')
plt.xlabel('Year')
plt.ylabel('Rate by Percentage')
plt.show()
```
